```python
import pandas as pd
pd.set_option('display.max_rows', 500) #change settings to output more
```

```python
beer_df = pd.read_csv('beer_data_cleaned7-19.csv')
beer_df.head() #import and check
```

```python
beer_df.describe() #basic descriptives
#average abv of 6% is a good sign that cleaning went well, as well as 37 ibu and 5.49 ph (the goal is 5.2-5.6 for conversion)
```

```python
beer_df.isnull().sum()
```

```python
beer_df.isnull().sum()/len(beer_df)*100
```

```python
threshold= (len(beer_df))*.01
print(threshold)
beer_df_trimmed = beer_df.groupby('Base Malt').filter(lambda x : len(x)>threshold)
beer_df_trimmed.head()
len(beer_df_trimmed)
#gets rid of base malts that are <= 1% of the total variation in base malts (using value counts)
```

```python
beer_df_trimmed = beer_df.groupby('hop1name').filter(lambda x : len(x)>threshold)
beer_df_trimmed.head()
len(beer_df_trimmed)
#gets rid of rows with hops that are <= 1% of the total variation in primary/hop 1 (using value counts)
```

```python
beer_df_trimmed.to_csv('beer_data_cleaned7-27.csv')
```
